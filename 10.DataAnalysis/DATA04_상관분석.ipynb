# 2. 상관계수

>## 1) 상관계수 개념

>## 2) 상관계수 종류

>>### (1) 피어슨 상관계수
- 두 변수간의 관계를 수량화하는 지표
- 1 : 양의 상관관계
- 0 : 상관없음
- -1 : 음의 상관관계

```python
import seaborn as sns
import pandas as pd

#데이터셋 불러오기
tips = sns.load_dataset('tips')

#팁과 식사금액 사이의 상관계수 계산
corr = tips[['total_bill', 'tip']].corr()

#산점도 및 회귀선 시각화
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
import seaborn as sns

#데이터셋 설정
tips = sns.load_dataset('tips')

#피어슨계수
corr = tips['total_bill'].corr(tips['tip'], method='pearson')

#값 출력
print('Pearson Correlation Coeffieicient: ', corr)
```

>>### (2) 공분산
- 상관 관계의 방향과 강도를 나타내는 지표
- 두 변수가 함께 변화하는 정도
- 양수 : 두 변수 같은 방향으로 변화
- 0 : 두 변수 독립적
- 음수 : 두 변수 다른 방향으로 변화


```python
import numpy as np 

x = [1,2,3,4,5]
y = [2,3,5,6,7]

cov_matrix = np.cov(x, y)

print(cov_matrix)
```

```python
import seaborn as sns
import pandas as pd

#데이터셋 불러오기
tips = sns.load_dataset('tips')

#공분산구하기
covariance = tips['total_bill'].cov(tips['tip'])
print('Covariance between total_bill and tip:', covariance)
```

>## 3) 상관관계 유형

>>### (1) 선형관계
- 두 변수 간 직선형태의 관계
- 상관계수 구할 수 있음

```python
import numpy as np
import matplotlib.pyplot as plt

#선형 관계 데이터 생성
x_linear = np.linspace(0, 10, 100)
y_linear = 2 * x_linear + 1

#선형 관계 시각화
plt.scatter(x_linear, y_linear)
plt.title('Linear relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

>>### (2) 비선형관계
- 두 변수 간 곡선형태의 관계
- 상관계수 측정 어려움
- 로지스틱 회귀, 다항회귀

```python
import numpy as np
import matplotlib.pyplot as plt

#비선형 관계 데이터 생성
x_nonlinear = np.linspace(-10, 10, 100)
y_nonlinear = x_nonlinear ** 2

# 비선형 관계 시각화
plt.scatter(x_nonlinear, y_nonlinear)
plt.title('Nonlinear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

# 4. 상관분석 종류

>## 1) 다중 상관 분석
- 독립변수 2개 이상, 1개의 종속변수 사이의 상관관계 분석
- 모든 변수간의 상관계수 행렬 구한 후 변수들 간의 상관관계 파악
- 회귀분석, 변수선택, 다중공선성 파악

```python
#타이타닉 다중상관분석

import seaborn as sns

#titanic 데이터셋에서 일부 변수 선택 
cols = ['survived', 'pclass', 'age', 'fare']

#선택한 변수로 새로운 데이터 프레임 생성
df = sns.load_dataset('titanic')[cols].dropna()

#다중 상관 분석 수행
corr = df.corr()

#결과 출력
print(corr)
```

>## 2) 부분 상관 분석
- 두 변수 간의 상관관계에서 다른 변수들의 영향 제거 후 얻어진 상관관계 계산하는 방법
- 편상관계수 이용 : 두 변수 간의 상관관계에서 다른 변수 영향 제거하고 나머지 변수만 계산한 상관관계

>## 3) 시계열 상관 분석
- 시간에 따라 변하는 두 변수 사이의 상관 관계 평가하는 통계적 기술
- 일별주가, 월별 판매량 등 시간에 따른 변화예측/동향파악에 유용함
- 자기회귀모델, 이동평균모델

```python
#삼성,LG전자 주식 데이터 활용 실습 (시간축 정보 필수)

import pandas as pd
import random 

#삼성 전자 주식데이터
dates = pd.date_range(start='2021-01-04', periods=100, freq='D')
samsung_prices = [random.randint(80000, 90000) for _ in range(100)]
samsung_data = {'Date':dates, '005930.KS': samsung_prices}
samsung_df = pd.DataFrame(samsung_data)
samsung_df.set_index('Date', inplace=True)

#LG 전자 주식데이터
lg_prices = [random.randint(140000, 160000) for _ in range(100)]
lg_data = {'Date': dates, '066570.KS': lg_prices}
lg_df = pd.DataFrame(lg_data)
lg_df.set_index('Date', inplace=True)

#두 데이터프레임을 합쳐서 하나의 데이터프레임 만들기
df = pd.concat([samsung_df, lg_df], axis=1)
df = df.loc[:, ['005930.KS', '066570.KS']]
df.columns = ['Samsung', 'LG']

#CSV파일로 저장
df.to_csv('stock_price.csv')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

#주식 데이터 불러오기
df = pd.read_csv('stock_price.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
print(df)

#두 종목의 수익률 계산
returns = df.pct_change()
print(returns)

#수익률 간의 상관 계수 계산
corr_matrix = returns.corr()
print(corr_matrix)
```

```python
# 히트맵그리기
sns.heatmap(corr_matrix, annot=True, cmap='rainbow')
plt.title('Stock Returns Correlation')
plt.show()
```
