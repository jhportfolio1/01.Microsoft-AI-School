>>>#### ④ 실습3 (중심경향성)
- 연속형 데이터의 중심 경향성을 파악하는 것 중요
- 평균(mean), 중앙값(median), 최빈값(mode) 계산 가능 
- 분포가 비대칭적인 경우 평균보다 중앙값/최빈값이 더 적합할 수 있음 

```python
import numpy as np
import matplotlib.pyplot as plt

#데이터생성
data = np.random.normal(loc=0, scale=1, size=10000)

#중심 경향성 계산
mean = np.mean(data)
median = np.median(data)
mode = np.round(np.mean(data))

#분포시각화
fig, ax = plt.subplots()
ax.hist(data, bins=50)
ax.axvline(mean, color='red', linestyle='dashed', linewidth=2)
ax.axvline(median, color='green', linestyle='dashed', linewidth=2)
ax.axvline(mode, color='purple', linestyle='dashed', linewidth=2)
ax.legend(['Mean: {:.2f}'.format(mean), 'Median: {:.2f}'.format(median), 'Mode: {}'.format(mode)])
ax.set_xlabel('Data Values')
ax.set_ylabel('Frequency')
ax.set_title('Data Distribution')
plt.show()
```

>>>#### ⑦ 실습6 (상관관계)
- 두 개의 지표간의 관계 
- 상관계수(Correlation Coefficient)를 이용해 파악 가능
- -1~1사이의 값으로 나타냄
- 양의 상관관계(1에가까움) : 변수 A증가시 B도 증가함
- 음의 상관관계(-1에가까움) : 변수 A증가시 B는 감소함

```python
import numpy as np 
import matplotlib.pyplot as plt

#데이터 생성
np.random.seed(123)
x = np.random.normal(loc=0, scale=1, size=100)
y = 2 * x + np.random.normal(loc=0, scale=1, size=100)

#선점도 시각화
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter Plot')
plt.show()

#상관관계 계산 (계수가 1에 가까울 수록 상관관계 높음)
corr = np.corrcoef(x, y)[0, 1]
print("상관계수:", corr)
```

>>>#### ⑧ 실습7 (수치형데이터/Q-Q플롯)
- Q : Qauntile(분위수) 의미, 분석하고자 하는 데이터 분포와 비교대상인 기준 분포의 분위수 값을 서로 대응시켜 그림
- 데이터 분위수와 기준 분포의 분위수 값을 계산하여 x,y축에 나타냄
- 두 데이터 분포가 유사하면 기울기 1과 같음

```python
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd

# load titanic dataset
titanic = sns.load_dataset('titanic')

# create Q-Q plot for age
stats.probplot(titanic['age'].dropna(), dist='norm', plot=plt)
plt.title('Q-Q plot for Age')
plt.show()
```

>>>#### ② 실습2 (상대빈도수)

```python
import numpy as np
import matplotlib.pyplot as plt

# 예시 데이터 생성
data = np.array([1,2,3,3,4,4,4,5,5,5,5])

# 데이터에서 각 값의 빈도수 계산
value_counts = np.bincount(data)
print(value_counts)
```

```python
# 각 값의 상대빈도수 계산

relative_frequencies = value_counts / len(data)
print(relative_frequencies)
```

```python
# 막대 그래프 시각화
plt.bar(np.arange(len(value_counts)), relative_frequencies)
plt.xticks(np.arange(len(value_counts)), np.arange(len(value_counts)))
plt.xlabel('Value')
plt.ylabel('Relative Frequency')
plt.title('Relative Frequency Bar Chart')
plt.show()
```

```python
from collections import Counter

def print_frequency(numbers):
 # 빈도수 계산
 counts = Counter(numbers)

 # (빈도수, 수) 쌍의 리스트 생성
 frequency_list = [(count, number) for number, count in counts.items()]

 # 빈도수를 기준으로 정렬
 sorted_list = sorted(frequency_list)

 # 결과 출력
 for count, number in sorted_list:
  print(f"수: {number}, 빈도수: {count}")

# 예시 입력 데이터
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

# 함수 호출
print_frequency(data)
```

>>### (3) 범주형 데이터

>>>#### ① 개념 및 기본실습
- 개념 : 정량적인 측정이 불가능하고 특정 카테고리에 속함
- 명목형 데이터(성별, 혈액형, 종교, 국적)와 순서형 데이터(학년, 만족도 수준, 경력)로 나뉘어짐
- 명목형 데이터 : 각 항목이 독립적이고 순서 나타내지 않음
- 순서형 데이터 : 항목 간의 순서가 있는 데이터(우수, 보통, 불합격, 옷사이즈 등) / 계층구조 나위어 있고 서열변수로 취급

>>>#### ② 범주형 데이터 분석 시 고려할 사항 (빈도수, 비율, 카이제곱 검정 등
- 빈도수 : 각 항목이 나타난 횟수 / 데이터 특성 파
- 비율 : 각 항복의 빈도수를 전체 데이터로 나눈 값
- 카이제곱 검정 : 두개 이상의 범주형 변수가 서로 관련있는지 검정함(성별과 직업의 상관관계) / 관찰된값과 기대값을 비교함
- 상관분석 : 범주형 변수 간의 상관관계 파악
- 로지스틱 회귀분석 : 종속변수와 연속형, 범주형 독립변수 간의 관계 파악 하기 위함

```python
# 카이제곱 검정 간단한 시각화 실습

import numpy as np
from scipy.stats import chi2_contingency

#예시 데이터 생성
observed_values = np.array([[10, 20, 30], [6,15,9]])

#카이제곱 검정 수행
chi2, p_value, dof, expected_values = chi2_contingency(observed_values)

#결과 출력
print(f"카이제곱 통계량: {chi2}")   #통계량 수치 높을수록 관련성 강해짐
print(f"p_value: {p_value}")        #p_value: 변수가 우연히 발생할 확률 / 작을수록 유의미함
print(f"자유도: {dof}")             #자유도: 검정에 사용딘 통계적 자유도의 수 
print("기대값:")                    #기대값: 특정한 조건에서 사건 발생가능성 평균값
print(expected_values)              #
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#데이터 불러오기
titanic_df = pd.read_csv('Titanic_data.csv')

#성별에 따른 생존률 계산
survived_by_sex = titanic_df.groupby('Sex')['Survived'].mean()

# 막대 그래프로 시각화
plt.figure(figsize=(6, 4))
sns.barplot(x=survived_by_sex.index, y=survived_by_sex.values)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#탑승 항구에 따른 생존자 수를 막대 그래프로 시각화
survived_emparked = titanic_df.groupby('Embarked')['Survived'].sum()
dead_embarked = titanic_df.groupby('Embarked')['Survived'].count() - survived_emparked

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(survived_emparked.index, survived_emparked.values, color='g', alpha=0.5, label='Survived')
plt.bar(dead_embarked.index, dead_embarked.values, bottom=survived_emparked.values, color='r',alpha=0.5, label='Dead')
plt.title('Survivors by Embarked')
plt.xlabel('Embarked')
plt.ylabel('Number of passengers')
plt.legend()
plt.show()
```

```python
# 객실등급에 따른 생존자 수
survived_class = titanic_df.groupby('Pclass')['Survived'].sum()
dead_class = titanic_df.groupby('Pclass')['Survived'].count() - survived_class

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(survived_class.index, survived_class.values, color='g', alpha=0.5, label='Survived')
plt.bar(dead_class.index, dead_class.values, bottom=survived_class.values, color='r', alpha=0.5, label='Dead')
plt.title('Survivors by Class')
```
