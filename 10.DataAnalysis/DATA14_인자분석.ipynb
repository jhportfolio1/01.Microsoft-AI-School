>## 2) 인자 분석의 가정

>>>### (1) 인자독립성
>>>- 인자들은 서로 독립적 (인자간 상관관계 없거나 매우 약함)

>>>- 상관관계 있다면 상관구조 반영하는 다른 분석방법 써야함 

>>>### (2) 인자의 정규성
>>>- 인자들이 정규분포 따른다는 가정 
>>>- 인자 분석의 통계적 기법 적용 위해 중요한 가정
>>>- 정규분포 따르지 않으면 비모수방법/변환기법 사용하면 됨

>>### (3) 선형성 가정
>>- 선형관계 기반으로 함.

# 8. 인자분석 - 고유값/고유벡터 해석

>## 1) 고유값
>- 고유값 : 인자분석에서 추출된 주성분의 중요도 나타냄
>- 각 주성분에 대해 고유값 할당되고, 고유값은 주성분 원본 데이터의 분산을 얼마나 잘 설명하는지 나타냄
>- 내림차순으로 정렬, 데이터의 변동성 더 잘 설명함
>- 고유값 크기 작을수록, 주성분 중요도 낮아짐 
>- 주성분 선택, 차원축소에서 고유값 크기 고려하여 결정 가능

>## 2) 고유벡터
>- 고유벡터 : 인자분석에서 추출된 주성분의 방향
>- 주성분에 대해 해당 주성분과 관련된 고유벡터가 있음
>- 주성분은 원본 변수간의 선형조합으로 이루어짐
>- 고유벡터는 해당 조합 계수를 나타냄
>- 고유벡터는 일반적 단위길이 가지고, 직교하는 성질로 독립적임

>- 원본 변수들이 특정 인자에 공통적으로 기여하는 경우, 해당 인자를 통해 원본 변수들 간의 관계 파악 가능

# 14. 인자분석 실습
- 랜덤데이터셋 사용


```python
import numpy as np
from sklearn.decomposition import FactorAnalysis

#데이터셋 생성
np.random.seed(0)
n_samples = 100   #샘플수
n_features = 5    #변수 수

#데이터셋 랜덤생성
X = np.random.randn(n_samples, n_features)

#인자분석수행
n_components = 2  #추출할 인자 수
factor_analysis = FactorAnalysis(n_components=n_components, random_state=0)
X_factors = factor_analysis.fit_transform(X)

#결과출력
print("원본 데이터셋 크기:", X.shape)
print("인자 분석 결과 (인자개수 = {}):".format(n_components))
print(X_factors)
```
