>>### (2) 쿼리편집기 창에 다음 쿼리 입력 후 실행

```python
SELECT TOP 20 pc.Name as CategoryName, p.name as ProductName
FROM SalesLT.ProductCategory pc
JOIN SalesLT.Product p
ON pc.productcategoryid = p.productcategoryid;
```
