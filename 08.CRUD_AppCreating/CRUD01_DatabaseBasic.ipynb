# 1. 데이터베이스 기초

>## 1) 데이터베이스란
- 데이터의 집합
- 데이터의 생성, 삭제, 업데이트 
- 엑셀/액세스도 하나의 데이터임
- 카카오/네이버 등의 대용량 데이터를 관리하려면 DBMS(Database Management System)가 필요함

>## 2) 엑셀 데이터
- Tab = Table
- cell = field
- 행 : Row
- 열 : column
- 관계형데이터베이스 : Tab 안의 정보를 다른 Tab에서 사용할 수 있음
- 정렬 및 검색이 가능

>## 3) 액세스
- SQL 등 사용해서 데이터 검색/쿼리 가능함

>## 4) DBMS(Database Management System)
- 관계형 데이터 베이스(RDBMS)를 통칭함
- 대표제품 : Oracle, MySQL(오라클인수), MariaDB, MsSqlServer, PostgreSQLimage.png
- Database Server : 공급자
- Client : 사용자
