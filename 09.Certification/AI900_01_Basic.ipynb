>## 1) AI란, 인간의 능력을 모방하는 소프트웨어

>## 2) 일반적인 인공지능 워크로드
- Machine Learning : 데이터 및 통계를 기반으로 한 예측 모델 - AI의 기초
- 이상 감지 : 비정상적인 패턴이나 이벤트를 감지하여 선제적 조치를 가능하게 하는 시스템
- Computer Vision : 카메라, 이미지 또는 비디오의 시각적 입력을 해석하는 애플리케이션
- 자연어처리 : 서면 또는 음성 언어를 해석하고 사용자와 대화에 참여할 수 있는 애플리케이션
- 정보 마이닝 : 데이터 원본에서 정보를 추출하여 검색 가능한 지식 저장소 만들기

>## 3) 책임있는 AI의 원칙과 해결과제
- 공정성 : 바이어스는 결과에 영향을 줄 수 있음 → 대출 승인 모델은 학습된 데이터의 바이어스로 성별에 따라 차별함
- 신뢰성 및 안정성 : 오류로 인해 해로울 수 있음 → 자율 주행 차량에서 시스템 오류가 발생하여 충돌이 발생
- 프라이버시 및 보안 : 데이터 노출가능 → 의료 진단봇이 중요한 환자 데이터로 학습되지만 데이터가 안전하지 않게 저장되어 있음
- 포용성 : 해결방법이 모든 사용자에게 작동하지 않을 수 있음 → 예측 앱은 시각 장애 사용자에게 오디오 출력 제공하지 않음
- 투명도 : 사용자는 복잡한 시스템을 신뢰해야 함 → AI기반 재무도구는 투자권장지침제공함, 이 권장사항은 무엇을 기준으로하는가?
- 책임 : AI 기반 결정은 누구의 책임인가? → 안면인식 증거에 따라 죄없는 사람이 범죄자로 확인되었음, 누구의 책임인가?

# 3. Azure ML

>## 1) 기계학습이란?
- 데이터에서 관계 찾아 예측 모델 만들기
- 샘플수집 -> 특징 -> 알고리즘으로 특징과 레이블간의 관계 찾음 -> 결과(캡슐화)

>## 2) 기계학습 유형

>>### (1) 감독되는 기계 학습
- 학습 데이터에는 알려진 레이블이 포함됨
- 회귀(레이블 숫자값) : 일/계절/날씨에 따른 자전거 대여 수 예측
- 분류(레이블이 분류/클래스) : 환자가 임상 측정값에 따라 당뇨병에 대한 위험에 처해있는지 여부 예측

>>### (2) 감독되지 않는 기계학습
- 학습데이터의 레이블은 지정되지 않음
- 클러스터링(유사한 항목 그룹화) : 배기가스 배출과 연비 특성이 유사한 차량은 클러스터로 구분

>## 3) 모델학습 및 유효성 검사
- 데이터를 학습 집합 / 유효성 검사 집합으로 분할
- 모델을 학습 데이터에 맞게 알고리즘에 적용
- 학습된 모델은 데이터의 관계를 캡슐화 함
- 모델을 사용하여 유효성 검사 데이터에서 예측 생성
- 평가 메트릭을 사용하여 예측 레이블과 실제 레이블 비교(감독됨) 또는 클러스터 분리측정(감독X) 반복

>## 4) 자동화된 기계학습 : 자동으로 ML 진행함
