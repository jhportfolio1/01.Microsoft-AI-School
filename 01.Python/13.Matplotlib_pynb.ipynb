## 4) 그래프 더 세세하게 그리기

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.sin(x-1))
plt.plot(x, np.sin(x-2))
plt.plot(x, np.sin(x-3))
plt.plot(x, np.sin(x-4))
plt.plot(x, np.sin(x-5))
```

### (1) 색상 정하기

```python
plt.plot(x, np.sin(x), color='blue')
plt.plot(x, np.sin(x-1), color='g')
plt.plot(x, np.sin(x-2), color='0.75')     #0~1은 흰색에서 검은색까지
plt.plot(x, np.sin(x-3), color='#FFDD44')
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3))
plt.plot(x, np.sin(x-5), color='chartreuse')
```

```python
plt.plot(x, x+0)
```

### (2) 선 모양 정하기

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

### (3) 컬러, 스타일 한번에 설정하기

```python
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.m')
plt.plot(x, x+3, ':b')
```

### (4) 세부조정하는법

- x, y축 상하 여백 조정 : 리미트 조정

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)        #x축 리미트를 조정한다.
plt.ylim(-1.5, 1.5)     #y축 리미트를 조정한다.
```

- 리미트 조정을 통한 상하반전

```python
plt.plot(x, np.sin(x))
plt.xlim(10,0)   
plt.ylim(1.2, -1.2)  
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')     #그래프 크기에 딱 맞게 표현하라(기본값)
```

### (4) 진짜 그래프 크기 보는법 : xy축을 1대1로 표시

```python
plt.plot(x, np.sin(x))
plt.axis('equal')        #1대1로 표시함
```

### (5) Label 붙이기

```python
plt.plot(x, np.sin(x))
plt.title('A sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.axis('equal')
```

```python
plt.plot(x, np.sin(x), '-g')
plt.plot(x, np.cos(x), ':k')
plt.axis('equal')
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':k', label='cos(x)')
plt.axis('equal')
```

### (8) plt.scatter 산점도를 표현


```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
color = rng.rand(100)
sizes = rng.rand(100) * 1000
plt.scatter(x, y, s=sizes,c=color, alpha=0.3)
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
color = rng.rand(100)
sizes = rng.rand(100) * 1000
plt.scatter(x, y, s=sizes,c=color, alpha=0.3, cmap='viridis')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0],features[1],c=iris.target, s=100, alpha=0.2, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
iris.feature_names[3]
```
