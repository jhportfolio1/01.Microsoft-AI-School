# 1. Cutout - Devries and Taylot, 2017
- Improved Regularization of Convolutional Neural Networks with Cutout
- Cutout : 이미지의 일부 영역을 임의 선택하여 검은색으로 채워서 이미지 정보를 지우는 것
- 모델이 해당 부분 무시하고 다른 부분을 더 잘 이해할 수 있음(성능개선)
- CIFAR-10, CIFAR100 데이터셋에서 Cutout 사용하여 ResNet, WARN 모델 성능 개선하는 실험 진행
- 결과 : Cutout 적용시 더 낮은 테스트 오차 보임

# 2. Mixup Zhang et al., 2018
- mixup: Beyond Empirical Risk Minimization
- 2개의 이미지 샘플과 레이블을 선형 결합하여 새로운 이미지 샘플 및 레이블 생성하는 방법
- 선형결합 가중치는 베타분포에서 무작위로 샘플링됨
- 생성된 이미지 샘플과 레이블은 기존 것을 결합한 것처럼 합성됨
- 이를 학습 모델에 사용하면 모델이 일반화를 더잘할수있도록 함
- 모델이 특정 클래스에 대해 과도하게 신뢰하지 않도록 도와줌

# 3. CutMix Yun et al.,2019
- 2019년 발표
- Mixup에서 영감받았지만 차이 있음
- 두개의 이미지 자르고 섞어서 합성 이미지 만듦
- 잘라낸 이미지 부분과 해당부분에 해당하는 레이블 섞어서 새로운 이미지 및 레이블 생성
- 장점 : 모델이 픽셀 레벨에서 다른 클래스 볼 수 있도록 함
- 모델이 세부적인 패턴 및 형상 인식하고 이에대한 불변성 개발하는데 도움됨
- 데이터증강이 일반화 성능 향상시키는데 도움된다고 입증함

# 4. AugMix Henddrycks el al.,2019
- AugMix : 다양한 이미지 강화 기술 조합, 신뢰성 일반성 향상에 중점둠
- 이미지 강화기술을 랜덤하게 섞어서 적용하는 것을 제안
- 이미지 강화기술로 이미지 강화 후 선형 조합 통해 랜덤한 강화이미지 생성
- 장점 : 데이터노이즈 및 과적합 줄이고 모델의 강건성 높이는데 성공
- 데이터셋에 대한 전반적 이해 필요하고 복잡한 구현이 필요해서 구현이 어령루 수 있음

# 5. AutoAugment(AA) Cubuk et al., 2019
- AutoAugment(AA) : 구글브레인에서 발표된 논문, 데이터 증강 기술 중 하나
- 데이터 증강을 자동으로 수행하여 이미지 분류 문제에서 SOTA(State of the Art) 성능 달성하는데 기여함
- 다양한 기법으로 이미지 변형함, 축소/좌우반전/회전
- 최적의 데이터 증강정책 찾기 위해 강화학습 사용
- 모델의 정확도 높은 정책 찾기 위해 대규모 데이터 세트에서 실험 수행
- 이렇게 찾은 정책은 다른 데이터셋에서도 잘 작동함
- 복잡하고 구현이 오래걸려서 전문적 연구 및 대규모 이미지 분류에만 사용되고 있음

# 6. Puzzle Mix Kim et al.,2020
- 이미지를 4개의 작은 패치로 나누어 재조합하는 방식으로 데이터 증강
- 각각 패치를 다른 이미지에서 무작위로 선택한 패치로 대체
- 생성된 4개의 패치를 다시 하나의 이미지로 조합하여 새로운 이미지 생성
- 데이터증강하면서 모델이 다양한 객체간 상관관계 학습할 수 있음
- 이미지내에서 객체간 상호작용 재현하며 새로운 상황 생성가능하여 증강 풍부해짐

# 7. Population Based Augmentation(PBA)
- 스탠포드 2019 논문
- AutoAugment에서 시작된 policy search 방식 개선 연구
- 기존 AA는 RL(강화학습)이용하여 정책찾았으나 오래걸리고 실험환경에 따라 정책이 자주 바뀌는 한계 있었음
- PBA : BO와 같은 최적화 기법 이용하여 정책 탐색
- 기존 AA에서 사용하는 RL에 비해 탐색속도 더빠르고 하이퍼파라미터 조정이 쉬움
- 훈련/검증/테스트로 데이터 분할한뒤 각각의 과정에서 정책탐색하고 업데이트

# 9. RandAugment(RA) Cubuk et al., 2020
- 간단하면서도 효과적인 데이터 증강 기법
- 2가지 하이퍼파라미터 사용
- 1번째 하이퍼파라미터 : N / N개의 데이터증강연산을 랜덤하게 선택
- 2번째 하이퍼파라미터 : M / 각 연산에 대해 강도 설정
- RA는 16개의 데이터 증강 연산 사용
- 가우시안 블러의 표준편차 = 0.5x5/10 = 0.25
- 컬러자극, 컨트라스트강화, 이미지평평화, 랜덤크롭

# 8. 용어정리

>## 1) BO
>- BO : Bayesian Optimization
>- 베이지안최적화의 줄임말, 함수의 최적화 문제를 해결하기 위해 사용되는 기술 중 하나
>- 기존의 그리드서치나 랜덤 탐색과 같은 방법보다 더 효율적으로 최적화 문제 해결 가능

>## 2) 사후분포
- 어떤 사건 및 인자에 대한 관측값 바탕으로 베이지안 정리 후 추정하는 분포
- 예시 ; 확률분포 생성 데이터집합에서 모수값에 대한 사전분포 정의후 사전분포와 데이터 이용하여 모수값 추정하는 과정을 통해 사후분포 구할 수 있음
- 추정한 모수값에 대한 불확실성 나타내며 모델의 예측 신뢰도 평가 가능
