>## 1) Bounding Box
>- 물체를 직사각형 모양 박스 안에 포함되도록 그리는 라벨링 방법
>- bbox : bbox 구성하는 대각선 관계의 두 점(x1, y1), (x2, y2)를 하나의 리스트에 넣거나, w/h를 한 리스트에 넣는 방식
>- [x1, y1, x2, y2] 또는 [x, y, w, h] 구성


>>### (1) Yolo Format Bounding Box
>>- bbox의 중간 위치의 좌표값을 사용
>>- [x_center, y_center, w, h]
>>- 좌표값은 이미지에 대한 상대적인 값을 사용
>>image.png
>>image.png

>## 2) Polygon
>- 다각형 모양으로 객체의 가시 영역 외곽선을 따라 점 찍어 그리는 라벨링 방법
>- 모델학습시 필요하여, 반드시 작성되어야 함
>- 이미지 내에서 인식대상인 객체의 위치 표시 위해 사용
>- [(x1, y1), (x2,y2), ... (xn, yn)] 또는 [ x1,y1,x2 ,y2, ... xn, yn ]
>- image.png

>## 3) Polyline

>## 2) XML
>- 데이터 저장/전송하는데 사용하는 마크업 언어
>- HTML 매우 유사
>- 데이터와 문서구조 설명하는데 중점을 둠
>- 텍스트 기반 형식, 태그/속성 제공
>- 데이터 = 요소(element)
>- 요소 : 속성(attribute), 값(value) 가짐
>- 데이터 계층 구조 구성 가능
>- 다양한 소프트웨어 시스템간 데이터 교환
>- <> - </> 단락 구성
>- node의 트리구소로 이루어짐
>image.png

```python
import os
import cv2
import matplotlib.pylab as plt
from xml.etree.ElementTree import parse

def xml_read(xml_path):
  root = parse(xml_path).getroot()

  #XML에서 모든 이미지 정보 가져오기
  image_info = root.findall('image')


  #이미지 정보에서 박스 정보 가져오기
  for image in image_info:
    bbox = image.findall('box')

    #이미지 사이즈 가져오기
    image_width = image.attrib['width']
    image_height = image.attrib['height']
    print("이미지크기 : ", image_width, image_height)

    #이미지 이름 가져오기
    image_name = image.attrib['name']
    print("이미지이름: ", image_name)
    #이미지 이름 가져올 패스 지정
    image_path = os.path.join('./', image_name)
    print("이미지경로: ", image_path)

    #이미지 가져오기
    image = cv2.imread(image_name)

    #image RGB 변환
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    #XML에서 좌표 가져오기
    for box_info in bbox:
      label = box_info.attrib['label']

      # 값 → string 타입
      xtl = box_info.attrib['xtl']
      ytl = box_info.attrib['ytl']
      xbr = box_info.attrib['xbr']
      ybr = box_info.attrib['ybr']

      #소수점 포함이라 바로 int값 변환 불가하여 float 변환
      xtl_f = float(xtl)
      ytl_f = float(ytl)
      xbr_f = float(xbr)
      ybr_f = float(ybr)

      #float → int 변환
      xtl_i = int(xtl_f)
      ytl_i = int(ytl_f)
      xbr_i = int(xbr_f)
      ybr_i = int(ybr_f)

      #시각화
      image = cv2.rectangle(image, (xtl_i, ytl_i), (xbr_i, ybr_i), (0,255,0), 2)

  plt.imshow(image)
  plt.show()



xml_read('annotations.xml')
```
