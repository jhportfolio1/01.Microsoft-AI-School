```python
### 1. qurey 선언 / chromedriver 실행
qurey = "사과"
service = Service("./chromedriver")
driver = webdriver.Chrome(service=service)

###2. query 검색창에 추가 
driver.get("https://www.google.co.kr/imghp?h1=ko")
# keyword = driver.find_element_by_xpath("/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea")
keyword = driver.find_element_by_class_name('gLFyf')
keyword.send_keys(qurey)
# /html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea -> 검색창
# /html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button -> 검색창 버튼 

### 검색창에 입력어 들어오면 검색 실행 
# driver.find_element_by_xpath("/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button/div/span/svg").click()
driver.find_element_by_class_name('Tg7LZd').send_keys(Keys.ENTER)

### 스크롤 내려가기 -> 검색 더보기 클릭 후 -> 스크롤 내려가기 
elem = driver.find_element_by_tag_name('body')
for i in range(60) : 
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.1)
    
try : 
    driver.find_element_by_class_name('mye4qd').send_keys(Keys.ENTER)
    for i in range(60) : 
        elem.send_keys(Keys.PAGE_DOWN)
        time.sleep(0.1)
except :
    pass

# 5. 이미지 개수 파악하기
# links = []
images = driver.find_elements(By.CSS_SELECTOR, value=".rg_i.Q4LuWd")

# for image in images : 
#     if image.get_attribute('data-src') !=None:
#         links.append(image.get_attribute('data-src'))
#     elif image.get_attribute('data-iurl') !=None:
#         links.append(image.get_attribute('data-iurl'))
#     elif image.get_attribute('src') !=None:
#         links.append(image.get_attribute('src'))
        
# print("찾은 이미지 개수 : " , len(links))

### 이미지 다운로드 
count = 0
for i in images : 
    start = time.time()
    
#     driver.find_elements(By.CSS_SELECTOR, value=".rg_i.Q4LuWd")

    
    os.makedirs("./download_image_data", exist_ok=True)
    # /html/body/div[2]/c-wiz/div[3]/div[2]/div[3]/div[2]/div/div[2]/div[2]/div[2]/c-wiz/div/div/div/div[3]/div[1]/a/img
    while True :
        try :
            i.click()
            # /html/body/div[2]/c-wiz/div[3]/div[2]/div[3]/div[2]/div/div[2]/div[2]/div[2]/c-wiz/div/div/div/div[3]/div[1]/a/img[1]
#             imgurl = driver.find_element_by_xpath('/html/body/div[2]/c-wiz/div[3]/div[2]/div[3]/div[2]/div/div[2]/div[2]/div[2]/c-wiz/div/div/div/div[3]/div[1]/a/img[1]').get_attribute("src")
            
            imgUrl = driver.find_element(By.XPATH, value="/html/body/div[2]/c-wiz/div[3]/div[2]/div[3]/div[2]/div/div[2]/div[2]/div[2]/c-wiz/div/div/div/div[3]/div[1]/a/img[1]").get_attribute("src")

            
            urllib.request.urlretrieve(imgurl, f"./download_image_data/{str(count)}_apple.png")
            print(f"{str(count + 1)} / {str(len(links))} / {qurey} / 다운로드 시간 : {str(time.time() - start)[:5]} 초")
            break
        except urllib.error.HTTPError as e : 
            print(f"HTTPError 발생 {e} : 재시도 중 .....")
            time.sleep(5)
        except Exception as e :
            print(f"Error 발생 {e} : 재시도 중 ......")
            time.sleep(5)
        
        if time.time() - start > 60 : 
            print("이미지 다운로드 실패 ...")
            break
    count = count + 1 
    
print("다운로드 완료")
driver.close()

```
