# 1. 쿠버네티스

>## 1) 개념
- 컨테이너를 관리하는 서비스
- 클라우드 환경에서 애플리케이션 배포, 운영 자동화 가능
- 대규모 컨테이너 자동으로 배포/배치/확장/장애복구 함

image.png

>## 2) 주요기능
- 오토스케일링 : 사용량에 따라 자동으로 컨테이너화된 앱과 리소스 크기를 조정하여 서버 안정성 조절
- 선언전 모델 : 선언적으로 서비스 관리하여 배포함 애플리케이션이 목적대로 실행되도록 함
- 자가복구 : 자동배치, 자동재시작, 자동복제, 자동확장 통해 애플리케이션 상태 확인과 자가 복구 수행
- 자동화된 롤아웃과 롤백 : 이전 버전으로 롤백, 일시중지, 배포 계속 기능 통해 구축/업데이트 자동화 가능
- 로드밸런싱 : 컨테이너 트래픽 너무 많거나 너무 적은 경우 자동으로 내외부 로드 밸런싱을 함

# 2. 실습 - Minikube 설치

>## 1) 사전 준비

>>### (1) 미니쿠베 다운로드
-  https://minikube.sigs.k8s.io/docs/start/

>>### (2) 최소사양
- CPU 2이상
- 메모리 2기가 이상(최소) / 12기가 이상이어야 원활
- 디스크 20기가 이상(최소) / 100기가 이상이어야 원활
- 가상화 툴 : Docker, Hyper-V

>## 2) 설치

- 파일 설치
https://storage.googleapis.com/minikube/releases/v1.22.0/minikube-linux-amd64


- cmd 설치

```python
sudo install minikube-linux-amd64 /usr/local/bin/minikube
```

```python
curl -LO https://dl.k8s.io/release/v1.22.1/bin/linux/amd64/kubectl
```

- 바이너리 사용할 수 있도록 권한과 위치 변경

```python
sudo install -o root -g root -m0755 kubectl /usr/local/bin/kubectl
```

- 정상 다운로드 확인

```python
minikube --help
```

- 버전 확인

```python
minikube version
```

>## 3) 미니쿠베 시작

```python
minikube start --driver=docker
```
