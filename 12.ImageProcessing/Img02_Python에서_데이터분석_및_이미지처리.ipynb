>## 2) 복수개체 저장 및 불러오기

```python
#numpy 데이터 불러오기

import numpy as np

#복수개체생성
array1 = np.arange(0,10)
array2 = np.arange(10,20)
print(array1, array2)
```

```python
#numpy 저장
np.savez("save.npz", array1=array1, array2=array2)

#numpy 불러오기
data = np.load('save.npz')

#객체 불러오기
result1 = data['array1']
result2 = data['array2']
print(result1, result2)
```

>>### (3) 원소의 정렬 - 각 열,행 기준으로 정렬

```python
array = np.array([[5, 10, 20, 30, 2], [7, 3, 5, 62, 6]])
org_array = array.copy()

#각 열을 기준으로 정렬 전
print("각 열을 기준으로 정렬 전 \n", array)

#각 열을 기준으로 정렬
array.sort(axis=0)
print("각 열을 기준으로 정렬 \n", array)
```

```python
#각 행을 기준으로 정렬 전
print("각 행을 기준으로 정렬 전 \n", org_array)

#각 행을 기준으로 정렬
org_array.sort(axis=1)
print("각 행을 기준으로 정렬 \n", org_array)
```

>>- 다른 모양(shape) 배열 간에 산술 연산 수행

>>### (11) Numpy 사용 이미지 저장시 효율성 비교

```python
pip install Pillow
```

```python
import numpy as np
from PIL import Image

#이미지 불러오기
img = Image.open('cat.png').convert('L')  #그레이스케일(회색조) 변경

#이미지 numpy 배열로 변환
img_array = np.array(img)
print(img_array)
```

```python
#numpy.save() 함수 사용해서 이미지 저장
np.save('lena_numpy', img_array)

print(img_array)
```

```python
#다시 PIL 이미지로 변환
img = Image.fromarray(np.uint8(img_array))
img.save('lena_numpy.png')

#Pil.Image.fromarray() 함수 사용하여 이미지 저장
img = Image.fromarray(np.uint8(img_array))
img.save('lena_pil.png')
```

```python
import os

#numpy.save()함수로 저장된 이미지 파일 크기 비교
np_save_szie = os.path.getsize('lena_numpy.npy')

#Pil.Image.fromarray() 함수로 저장된 이미지 파일 크기 비교
pil_save_size = os.path.getsize('lena_numpy.png')
```

```python
print(f'numpy.save() 함수로 저장한 파일 크기 : {np_save_szie} bytes')
print(f'\n PiL.Image.fromarray() 함수로 저장한 파일 크기 : {pil_save_size} bytes')
```
