>>### (3) Dot 연산 : 벡터의 곱
- 같은 차원의 각 성분끼리의 곱을 모두 더하는 계산
- 최종값이 1개이기 때문에 0차원의 스칼라가 나타남

image.png

```python
x = np.array([1,2,0])
y = np.array([0,2,0])
z = np.dot(x, y)

print(z)
print(z.shape)
print(z.ndim)
```

>>### (1) 2차원 텐서 간의 연산

```python
a = np.array([[1,2],[3,4]])
b = np.array([[10,10],[10,10]])

print(a * b)
```

>>### (2) 전치행렬 (행과 열을 바꿈)
- 데이터 가로, 세로가 바뀌어 있을때 이용하면 좋음

```python
a = np.array([[1,2,3],[4,5,6]])

print(a)
print(a.shape)
```

```python
a_ = a.T #전치행렬

print(a_)
print(a_.shape)
```
