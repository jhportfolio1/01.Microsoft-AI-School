# 2. 신경망 기초 - 딥러닝을 위한 수학

- 수학 패키지

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

>## 1) 일차함수 : y=ax+b
image.png

```python
def linear_function(x):
  a=0.5
  b=2

  return a*x+b
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5,5,0.1)
y = linear_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

>## 2) 이차함수 : y=ax2 + bx + c
image.png

```python
def quadratic_fuction(x):
  a = 1
  b = -1
  c = -2

  return a*x**2 + b*x + c
```

```python
print(quadratic_fuction(2))
```

```python
x = np.arange(-5,5,0.1)
y = quadratic_fuction(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
```

>## 3) 삼차함수(다항함수) : ax3 + bx2 + cx + d
image.png

```python
def cublic_function(x):
  a = 4
  b = 0
  c = -1
  d = -8

  return a*x**3 + b*x**2 + c*x + d
```

```python
x = np.arange(-5,5,0.1)
y = cubic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cublic Function')
```
